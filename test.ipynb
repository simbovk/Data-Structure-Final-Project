```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Reading Files

```python
size = []
gaussian_exe = []
uniform_exe = []

with open('C:/Users/Lenovo/Desktop/DSproject/inputs/samples.txt') as my_file:
    for line in my_file:
        splitted = line.split()
        size.append(splitted[0])
        gaussian_exe.append(splitted[2])
        uniform_exe.append(splitted[1])
```

# Result Table

```python
dic = {'gaussian' : gaussian_exe, 'uniform' : uniform_exe}
df = pd.DataFrame(dic,index=size,columns=['gaussian', 'uniform'])
df
```

# Plot

```python
sns.set_theme()
fig, ax = plt.subplots(2, 1, figsize = (15, 10))
ax[0].plot(size , gaussian_exe, label = 'gaussian contribution',color = 'blue')
ax[1].plot(size , uniform_exe, label = 'uniform contribution', color = 'red')
ax[0].set_xlabel('size of inputs')
ax[1].set_xlabel('size of inputs')
ax[1].set_ylabel('time(s)')
ax[0].set_ylabel('time(s)')
ax[0].invert_xaxis()
ax[0].invert_yaxis()
ax[1].invert_xaxis()
ax[1].invert_yaxis()
ax[0].legend()
ax[1].legend()
sns.set_style('darkgrid')
```

# Scatter Plot 

```python
sns.set_theme()
fig, ax = plt.subplots(2, 1, figsize = (15, 10))
ax[0].scatter(size , gaussian_exe, label = 'gaussian contribution',color = 'blue')
ax[1].scatter(size , uniform_exe, label = 'uniform contribution', color = 'red')
ax[0].set_xlabel('size of inputs')
ax[1].set_xlabel('size of inputs')
ax[1].set_ylabel('time(s)')
ax[0].set_ylabel('time(s)')
ax[0].invert_xaxis()
ax[0].invert_yaxis()
ax[1].invert_xaxis()
ax[1].invert_yaxis()
ax[0].legend()
ax[1].legend()
sns.set_style('darkgrid')
```
